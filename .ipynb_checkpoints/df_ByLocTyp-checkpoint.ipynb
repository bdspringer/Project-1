```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File path to listings.csv
listings_csv_path = "Data Files for AirBnB Project/listings.csv"

#read listings.csv into a database
listings_df = pd.read_csv(listings_csv_path)
listings_df.head()
```

```python
###  delete rows with o days availability  ###

#delete rows that have 0 annual avalability
listings_df = listings_df[listings_df.availability_365 != 0]
print(f"{len(listings_df)} properties remain")
```

```python
###  drop unecessary columns  ###
del listings_df["name"]
del listings_df['host_id']
del listings_df['host_name']
del listings_df['neighbourhood_group']
listings_df.head()
```

```python
### group by neighborhood and room type - find median price / avg number of days 
###  by neighborhood and room type
rmtype_dfg = listings_df.groupby(["neighbourhood","room_type"]).agg({"price":"median", "calculated_host_listings_count":"mean"})
#rename columns
rmtype_dfg.rename(columns={'price': 'median price', 'calculated_host_listings_count':'average days rented'}, inplace=True)

rmtype_dfg
```

```python
###   look only at Entire home or apartments sorted by max(median price)
house_dfg = rmtype_dfg[rmtype_dfg.index.map(lambda x: x[1] == ("Entire home/apt"))]
house_dfg = house_dfg.sort_values(["median price"], ascending=False) 
house_dfg
```
